```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.webdriver.chrome.options import Options as ChromeOptions


options = ChromeOptions()
#user_agent = "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_12_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/61.0.3163.100 Safari/537.36"
#options.add_argument('user-agent=' + user_agent)
options.add_argument("--chromedriver=/content/drive/MyDrive/chromedriver.exe")
options.add_argument("lang=ko_KR")
options.add_argument('headless')
options.add_argument('window-size=1920x1080')
options.add_argument("disable-gpu")
options.add_argument("--no-sandbox")

# chrome driver
driver = webdriver.Chrome(options=options) # <- options로 변경
```

```python
import pandas as pd
from selenium.webdriver.common.by import By
import time
```

```python
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.wait import WebDriverWait
```

```python
#exc파일 읽어오기
exc = pd.read_excel('excccc.xlsx')

description=[]
urll=[]
```

```python
exc.head()
```

```python
url = "https://delight.duksung.ac.kr/"
driver.get(url)

#for문을 이용해 입력할 텍스트 변수 지정
for i in range(len(exc.index)):
    #print(exc.iloc[i, 0])
    name = exc.iloc[i, 0]

    # 돋보기 버튼을 클릭하는 명령어
    a = driver.find_element(By.XPATH, '/html/body/div[2]/header/div/nav/ul/li[7]/button') #돋보기버튼
    driver.execute_script("arguments[0].click();", a)

    # 검색 박스를 클릭하는 명령어
    b = driver.find_element(By.XPATH, '/html/body/div[2]/header/form/div/div[1]/input') #검색박스
    driver.execute_script("arguments[0].click();", b)

    time.sleep(1)

    # 검색할 내용 입력하는 명령어
    #input = driver.find_element(By.Tag_Name, '//*[@id="ModuleEcoProgramListForm"]/div[1]/input');
    b.send_keys(name)

    # 검색 버튼을 누르는 명령어
    c = driver.find_element(By.XPATH, '/html/body/div[2]/header/form/div/div[1]/button[1]') #검색돋보기버튼
    c.click()
    #driver.execute_script("arguments[0].click();", c)

    # 상세 프로그램 클릭 명령어
    link_selector = "body > div:nth-child(2) > main > div > div > ul.columns-4 > li:nth-child(1) > div > a"
    links = driver.find_element(By.CSS_SELECTOR, link_selector) #긁어올 텍스트 위치

    url_2 = links.get_attribute("href")
    #print(url_2)
    driver.get(url_2)

    time.sleep(1)

    # 텍스트 긁어오기
    f = driver.find_element(By.XPATH, '//*[@id="ModuleEcoProgramView"]/div[1]/div/div[3]/div[2]/div/div[1]') #긁어올 텍스트 위치
    text = f.text
    description.append(text)
    urll.append(url_2)
```
